# Exploratory Data Analysis

## Age Group

```python
# ORDERS BY AGE GROUP

graph = sns.countplot(x='Age Group', data=df, hue='Age Group')
sns.set(rc = {'figure.figsize':(10,5)})
plt.show()
```

```python
ag_sales = df.groupby(['Age Group'],as_index = False) ['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc = {'figure.figsize':(10,5)})
sns.barplot(data = ag_sales, x ='Age Group', y = 'Amount', hue ='Age Group')
plt.show()
```

```python
# From Above Graph Most Of The Buyers are Between 26-35 Age.
```

## Order by Age-Group And Gender

```python
df.columns
```

```python
gen = sns.countplot(x= 'Age Group', data=df, hue='Gender')
sns.set(rc = {'figure.figsize':(8,5)})

for bars in gen.containers:
    gen.bar_label(bars)

plt.show()
```

```python
# From Above Graph Where Most Of Buyers Are Females And The Age Group is Between 26-35.
```

## State

```python
sta = sns.countplot(x = 'State', data = df, hue='State')
sns.set(rc = {'figure.figsize':(18,5)})
for bars in sta.containers:
    sta.bar_label(bars)
plt.show()
```

```python
top_sta = df.groupby(['State'], as_index = False) ['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(17,5)})
sns.barplot(data = top_sta, x = 'State', y = 'Orders', hue = 'State')
plt.show()
```

```python
# "The first graph displays the number of orders per state."

# "The second graph displays the top 10 states by total Orders."
```

## Occupation 

```python
occ = sns.countplot(data = df, x = 'Occupation', hue = 'Occupation')
for bars in occ.containers:
    occ.bar_label(bars)
plt.show()
```

```python
occ_amt = df.groupby(['Occupation'],as_index = False) ['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = occ_amt, x = 'Occupation', y = 'Amount', hue = 'Occupation')
plt.show()
```

```python
# 1 Graph of Occupation :- The first graph illustrates sales distribution by occupation
# 2 Graph of Occupation :- The second graph show that the most of buyers belong to the IT, healthcare, and aviation sectors."
```

## Product_Category

```python
pd = sns.countplot(x = 'Product_Category', data = df, hue='Product_Category')
sns.set(rc={'figure.figsize':(25,7)})
for bars in pd.containers:
    pd.bar_label(bars)

plt.show()
```

```python
pd_sales = df.groupby(['Product_Category'],as_index = False) ['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.barplot(data = pd_sales, x = 'Product_Category', y = 'Amount', hue = 'Product_Category')
sns.set(rc={'figure.figsize':(25,5)})
plt.show()
```

```python
# 1 Graph of Product Category :- The first graph shows the sales distribution across product categories.
# 2 Graph of Product Category :- The second graph indicates that the highest sales are in the Food, Clothing,and Apparel categories.
```

## Sales By State

```python
st = sns.countplot(data = df , x ='State', hue = 'State')
for bars in st.containers:
    st.bar_label(bars)
plt.show()
```

```python
sta_amt = df.groupby(['State'],as_index = False) ['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.barplot(data = sta_amt, x = 'State', y = 'Amount', hue = 'State')
plt.show()
```
